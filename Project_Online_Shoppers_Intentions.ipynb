**â¢ Unbalanced dataset & Resampling strategy**

Let's first check how unbalanced our dataset is.

````python
Y_count = y.value_counts()
print('Class False:', Y_count[0])
print('Class True :', Y_count[1])
print('Proportion:', round(Y_count[0] / Y_count[1], 2), ': 1')
colors =["#5ec0ca", "#6287a2"]

fig,ax = plt.subplots(figsize=(10,7))

Y_count.plot(kind='pie',title='Revenue class proportion', autopct='%1.0f%%',colors=colors);
````

As we can see, our dataset is highly unbalanced. Oversampling the minority class is one method for dealing with umbalanced datasets. The most basic method involves duplicating examples from the minority class, even though these examples add no new information to the model. Instead, new examples can be created by combining existing ones. The Synthetic Minority Oversampling Technique, or SMOTE, is a type of data augmentation for the minority class.

To create a synthetic individual, the steps defined in the SMOTE algorithm are :

1. Randomly select an âinitialâ minority observation.
2. Identify its closest neighbors k among the minority observations (where k is a user-defined parameter).
3. Randomly choose one of the nearest neighbors k.
4. Randomly generate an Î± ranged as follow  <code> 0 < Î± < 1 </code>
5. Create a new individual between the initial observation and the nearest selected neighbour, based on the coefficient value. For example, if <code> Î± = 0.5 </code>, the new individual will be positioned halfway between the initial observation and the nearest selected neighbour

````python
!pip install imblearn
````

````python
from imblearn.over_sampling import SMOTE
from matplotlib.pyplot import *

smote = SMOTE(sampling_strategy='minority')
X, y = smote.fit_resample(X, y)
````

````python
Y_count = y.value_counts()
print('Class False:', Y_count[0])
print('Class True :', Y_count[1])
print('Proportion:', round(Y_count[0] / Y_count[1], 2), ': 1')
colors =["#5ec0ca", "#6287a2"]

fig,ax = plt.subplots(figsize=(10,7))

Y_count.plot(kind='pie',title='Revenue class proportion', autopct='%1.0f%%',colors=colors);
````

As shown by the above pie, now we have a balanced dataset.

## 5. Machine Learning & Tuning
---

In order to get the best set of hyperparameters we are going to use the  Grid Search method. <br>

Grid search is a model hyperparameter optimization technique. In scikit-learn, this technique is provided in the GridSearchCV class. <br>

<li>When constructing this class, we must provide the list of hyperparameters to evaluate in the <code> param_grid </code> argument. </li>

<li>We will be using  the metric <code> f1 </code>, as it combines the metrics precision and recall.</li>

<li>By default, the grid search will only use one thread. </li>
<li>By setting the <code> n_jobs </code> argument in the <code> GridSearchCV </code> constructor to -1, the process will use all cores on our machine.</li> 

The GridSearchCV process will then construct and evaluate one model for each combination of parameters. Cross validation is used to evaluate each individual model.
Grid Search passes all combinations of hyperparameters one by one into a model and check the result. Finally it gives us the set of hyperparemeters which gives the best result after passing in the model. <br>

The models on which we are going to perform this process are the following : KNN, Naive Bayes Classifier, Decision Tree, Random Forest, XGBoost, SVC, Stochastic Gradient Descent and Gradient Boosting Score. <br>

For each model, we will stick to the following steps : <br>
1. Import the necessary libraries <br>
2. Define the grid parameters <br>
3. Apply <code> GridSearchCV </code> and evaluate the result <br> 
4. Save the best model obtained in a specified variable using <code> best_estimator_ </code>
5. Save the score obtained on the test set using the model we just defined earlier.
6. Plot the <code> confusion matrix </code> to evaluate the performance of the model.

 **â¢ KNN**

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
````

````python
params = {'n_neighbors' : np.arange(2,20), 
          'metric':['euclidean','manhattan','minkowski','chebyshev'],
          'weights': ('uniform', 'distance'),
         }
knn_grid = GridSearchCV(KNeighborsClassifier(),param_grid=params,cv=5)

knn_grid.fit(X_train,Y_train)

print('Best Parameters : {}'.format(knn_grid.best_params_))
print('Best Accuracy Through Grid Search : {:.3f}\n'.format(knn_grid.best_score_))
````

````python
model_knn = knn_grid.best_estimator_
````

````python
test_score_knn = model_knn.score(X_test,Y_test)
test_score_knn
````

````python
from sklearn.metrics import plot_confusion_matrix

fig, ax = plt.subplots(figsize=(8, 8))
plot_confusion_matrix(model_knn, X_test, Y_test, display_labels=['Not Purchased',
 'Purchased',], cmap=plt.cm.Blues,ax=ax);
````

````python
model_nb = bernoulli_nb_grid.best_estimator_
````
